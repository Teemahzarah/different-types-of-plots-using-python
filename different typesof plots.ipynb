```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
```

```python
tip =sns.load_dataset('tips')
flight=sns.load_dataset('flights')
tip.head()
```

```python
tipscor =tip.corr()
sns.heatmap(tipscor)
```

```python
tipscor
```

```python
flightpi =flight.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(flightpi,cmap='coolwarm',linecolor='white',linewidths=0.1)
```

```python
sns.clustermap(flightpi,standard_scale=100)
```

## grid

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
grid =sns.PairGrid(iris)
grid.map_diag(sns.distplot)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
```
